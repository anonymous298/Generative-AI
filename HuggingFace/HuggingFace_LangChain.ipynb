# Using HuggingFaceEndpoint

### For using Models as and API from HuggingFaceHub

```python
from google.colab import userdata
sec_key = userdata.get('HUGGINGFACEHUB_API_TOKEN')
```

```python
import os
os.environ["HUGGINGFACEHUB_API_TOKEN"] = sec_key
```

```python
from langchain_huggingface import HuggingFaceEndpoint
```

```python
repo_id = "mistralai/Mistral-7B-Instruct-v0.3"

llm = HuggingFaceEndpoint(repo_id=repo_id, temperature=0.7, max_length=128, token=sec_key)
```

```python
llm.invoke('What is AI?')
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

template = """
Question: {question}
Answer: define it with first principles
"""

prompt = PromptTemplate(template=template, input_variables=["question"])

llm_chain = LLMChain(llm=llm, prompt=prompt)
print(llm_chain)
```

```python
response = llm_chain.invoke('How Can I be a Expert coder?')
print(response)
```

```python
response.get('text')
```

# Using HuggingFacePipeline

### For loading model locally on our ram

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
model = AutoModelForCausalLM.from_pretrained('openai-community/gpt2')
tokenizer = AutoTokenizer.from_pretrained('openai-community/gpt2')
```

```python
pipe = pipeline('text-generation', model=model, tokenizer=tokenizer, max_new_tokens=100)
```

```python
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
llm.invoke('blockchain')
```

```python
prompt_template = PromptTemplate(input_variables=['question'], template='Question: {question} Answer: first principles')

chain = LLMChain(llm=llm, prompt=prompt_template)
```

```python
chain.invoke('Human is')
```

```python
gpu_llm = HuggingFacePipeline.from_model_id(
    model_id='gpt2',
    task='text-generation',
    device=0,
    pipeline_kwargs={'max_new_tokens': 100}
)
```

```python
template = """Question: {question}
Answer: Let's think step by step."""

prompt = PromptTemplate(input_variables=['question'], template=template)

# chain = LLMChain(llm=gpu_llm, prompt=prompt)
chain = prompt | llm
```

```python
chain.invoke({'question' : 'What is AI?'})
```
